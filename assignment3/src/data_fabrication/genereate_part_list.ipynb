# Generate CPU part list

```python
from bs4 import BeautifulSoup, NavigableString
from pandas import Series, DataFrame
import pandas as pd
import random
```

```python
cpu_list = open('html/CPU_list.html')
soup = BeautifulSoup(cpu_list, 'lxml')

cpu_list = soup.select('div.td__nameWrapper p')
cpu_name_list = []
for cpu in cpu_list :
    cpu_name_list.append(cpu.text)
    
cpu_list = soup.select('td.td__spec--1')
cpu_coreCount_list = []
for core in cpu_list :
    inner = [element for element in core if isinstance(element, NavigableString)]
    text = inner[0]
    cpu_coreCount_list.append(text)
    
cpu_list = soup.select('td.td__spec--2')
cpu_clockSpeed_list = []
for clock in cpu_list :
    inner = [element for element in clock if isinstance(element, NavigableString)]
    text = inner[0][:len(inner[0]) - 4]
    cpu_clockSpeed_list.append(text)
    
cpu_list = soup.select('td.td__spec--4')
cpu_TDP_list = []
for tdp in cpu_list :
    inner = [element for element in tdp if isinstance(element, NavigableString)]
    text = inner[0][:len(inner[0]) - 2]
    cpu_TDP_list.append(text)
    
cpu_list = soup.select('td.td__price')
cpu_price_list = []
for price in cpu_list :
    inner = [element for element in price if isinstance(element, NavigableString)]
    text = inner[0]
    cpu_price_list.append(text)
    

data = {'partNo': list(range(10000, 10000 + len(cpu_coreCount_list))) ,'name': cpu_name_list, 'coreCount': cpu_coreCount_list, 'clockSpeed': cpu_clockSpeed_list, 'TDP': cpu_TDP_list}

frame = DataFrame(data)

frame.head()

# frame.to_csv('csv/cpu_list.csv', index=False)
```

# Generate Memory part list

```python
mem_list = open('html/Memory_list.html')
soup = BeautifulSoup(mem_list, 'lxml')

mem_list = soup.select('div.td__nameWrapper p')
mem_name_list = []
for mem in mem_list :
    mem_name_list.append(mem.text)
    
mem_list = soup.select('td.td__spec--1')
mem_speed_list = []
for speed in mem_list :
    inner = [element for element in speed if isinstance(element, NavigableString)]
    text = inner[0][5:]
    mem_speed_list.append(text)
    
mem_list = soup.select('td.td__spec--3')
mem_capacity_list = []
mem_sticks_list = []
for mem in mem_list :
    inner = [element for element in mem if isinstance(element, NavigableString)]
    text = inner[0].split()
    text[2] = text[2][:len(text[2]) - 2]
    capacity = (int)(text[0]) * (int)(text[2])
    sticks = (int)(text[0])
    mem_capacity_list.append(capacity)
    mem_sticks_list.append(sticks)
    
mem_list = soup.select('td.td__price')
mem_price_list = []
for price in mem_list :
    inner = [element for element in price if isinstance(element, NavigableString)]
    text = inner[0]
    mem_price_list.append(text)
    
data = {'partNo': list(range(10300, 10300 + len(mem_name_list))) ,'name': mem_name_list, 'speed': mem_speed_list, 'capacity': mem_capacity_list, 'sticks': mem_sticks_list}

frame = DataFrame(data)

frame.head()

# frame.to_csv('csv/memory_list.csv', index=False)
```

# Generate Storage part list

```python
storage_list = open('html/Storage_list.html')
soup = BeautifulSoup(storage_list, 'lxml')

storage_list = soup.select('div.td__nameWrapper p')
storage_name_list = []
for storage in storage_list :
    storage_name_list.append(storage.text)
    
storage_list = soup.select('td.td__spec--1')
storage_capacity_list = []
for cap in storage_list :
    inner = [element for element in cap if isinstance(element, NavigableString)]
    text = inner[0]
    if text[-2:] == 'TB' :
        text = (int)((float)(text[:len(text) - 3]) * 1024)
    else :
        text = (int)(text[:len(text) - 3])
    storage_capacity_list.append(text)
    
storage_list = soup.select('td.td__spec--3')
storage_type_list = []
for stype in storage_list :
    inner = [element for element in stype if isinstance(element, NavigableString)]
    text = inner[0]
    if (text != 'SSD') :
        text = 'HDD'
    storage_type_list.append(text)
    
storage_list = soup.select('td.td__price')
storage_price_list = []
for price in storage_list :
    inner = [element for element in price if isinstance(element, NavigableString)]
    text = inner[0]
    storage_price_list.append(text)
    
data = {'partNo': list(range(10400, 10400 + len(storage_name_list))) ,'name': storage_name_list, 'type': storage_type_list, 'capacity': storage_capacity_list}

frame = DataFrame(data)

frame.head()

# frame.to_csv('csv/storage_list.csv', index=False)
```
