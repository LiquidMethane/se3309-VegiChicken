```python
# Define, navigate, soup, and look at it
url = "https://pcpartpicker.com/products/cpu/fetch"
data = requests.get(url).text
cpuSoup = BeautifulSoup(data,features="html.parser")
cpuSoup
```

```python
# From here we actually see that it is an API structure
# So it seems like it may just be a request
# Let's test this hypothesis
```

```python
# Yes it is!
cpuData = requests.get(url).json()
cpuData
```

```python
# Let's go straight for the resulting html
cpuData['result']['html']
```

```python
# However, we see that it is actually a string
# So now let's use Beautiful Soup on it
cpuSoup = BeautifulSoup(cpuData['result']['html'],features="html.parser")
```

```python
# Let's separate each product like they did using tr
productContainer = cpuSoup.findAll("tr",{"class":"tr__product"})
```

```python
# Now we have each product separated
# And taking a look at the structure of each product:
productContainer[0]
# We also see that its separated quite nicely
```

```python
# Now let's get someparts that we might want 
# like url for each item
for a in productContainer[0].find_all("a",href=True):
    url = "https://pcpartpicker.com"+ a['href']
```
