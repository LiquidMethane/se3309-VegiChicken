```python
from bs4 import BeautifulSoup, NavigableString
from pandas import Series, DataFrame
import pandas as pd
import random
```

```python
random_name_list = open('html/random_name_list.html')
soup = BeautifulSoup(random_name_list, 'lxml')

random_name_list = soup.select('ol#result li div span')

name_list = []
for name in random_name_list :
    name_list.append(name.text)
```

```python
name_list
```

```python
for a in range(0, len(name_list)) :
    for b in range(a + 1, len(name_list)) :
        if name_list[a] == name_list[b] :
            print('identical pass detected')
            break
print('passes are good')
```

```python
file = open('txt/email_domain.txt')

domain_list = file.read().split()
```

```python
domain_list
```

```python
email_list = []

for name in name_list :
    nameArr = name.split();
    email_list.append(nameArr[0].capitalize() + nameArr[1].capitalize() + (str)(random.randint(1000, 9999)) + '@' + domain_list[random.randint(0, len(domain_list) - 1)])
```

```python
email_list
```

```python
frame = DataFrame({'email': email_list})
frame.to_csv('csv/email_list.csv')
frame.head()
```
