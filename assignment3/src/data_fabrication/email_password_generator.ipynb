```python
file = open('txt/pass.txt')
s = file.read()
pass_list = s.splitlines()
pass_list
```

```python
for a in range(0, len(pass_list)) :
    for b in range(a + 1, len(pass_list)) :
        if pass_list[a] == pass_list[b] :
            print('identical pass detected')
            break
print('passes are good')
```

```python
df['password'] = pass_list

df.head()
```

```python
file = open('txt/nouns.txt')
s = file.read()
nouns = s.splitlines()
nouns
```

```python
file = open('txt/adjectives.txt')
s = file.read()
adj = s.splitlines()
adj
```

```python
file = open('txt/adverbs.txt')
s = file.read()
adv = s.splitlines()
adv
```

```python
import random

nickname_list = []

for i in range(0, 1000) :
    d = 'qqq'
    
    while len(d) > 20 or len(d) < 12:
        a = adv[random.randint(0, len(adv) - 1)].capitalize()
        b = adj[random.randint(0, len(adj) - 1)].capitalize()
        c = nouns[random.randint(0, len(nouns) - 1)].capitalize()
        d = a + b + c
    nickname_list.append(d)
```

```python
nickname_list
```

```python
df['password'] = nickname_list
```

```python
df.head()
```

```python
username_list = []
print(len(adj))
print(len(nouns))

for i in range(0, 1000) :
    d = 'qqqqqqqqqqq'
    
    while len(d) > 8 :
        b = adj[random.randint(0, len(adj) - 1)].capitalize()
        c = nouns[random.randint(0, len(nouns) - 1)].capitalize()
        d = b + c
    username_list.append(d)
```

```python
username_list
```

```python
df['nickname'] = username_list
```
